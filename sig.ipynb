```python
# Criando DataFrame a partir de Dicionário

dados = {
    'nome': ['pessoa 1','pessoa 2', 'pessoa 3'],
    'idade': [18, 20, 22],
    'cidade': ['Porto Alegre', 'Porto Alegre', 'Porto Alegre']}

df = pd.DataFrame(dados)
df
```

```python
# A partir de listas

nome = ['pessoa 1','pessoa 2', 'pessoa 3']
idade = [18, 20, 22]
cidade = ['Porto Alegre', 'Porto Alegre', 'Porto Alegre']

label = ['nome', 'idade', 'cidade']
lista_colunas = [nome, idade, cidade]

dados = dict(list(zip(label, lista_colunas)))

df = pd.DataFrame(dados)

df.head()
```
